```python
import os
import numpy as np
import cv2
import tensorflow
from tensorflow import keras
import keras.layers
from  keras.models import load_model
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt
```

```python
folder="/content/drive/MyDrive/DL/CNN/Models"
filename='face-mask.h5'
filepath = os.path.join(folder,filename)
model = load_model(filepath)
```

```python
def result():
    path = input()
    img= cv2.imread(path)
    cv2_imshow(img)
    img = cv2.resize(img,(128,128))
    img= img/255

    input1= np.reshape(img,[1,128,128,3])

    y_pred=model.predict(input1)
    print(y_pred)
    if(y_pred>0.5):
        print("Mask")
    else :
        print("NO mask")
    return
```
